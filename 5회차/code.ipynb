```python
from ultralytics import YOLO
import cv2
import os

model = YOLO("yolo11n-pose.pt")


def is_video_source(source):

    if isinstance(source, int) or (isinstance(source, str) and source.isdigit()):
        return True
    if isinstance(source, str):
        ext = os.path.splitext(source)[1].lower()
        return ext in {".mp4", ".avi", ".mov", ".mkv", ".webm"} or source.startswith(("rtsp://", "rtmp://"))
    return False

def show_pose_window(source, conf=0.50, device=None):
    win_name = "Pose Estimation"
    cv2.namedWindow(win_name, cv2.WINDOW_NORMAL)

    streaming = is_video_source(source)
    last_frame = None

    for result in model.predict(source=source, stream=True, conf=conf, device=device):
        frame = result.plot()
        last_frame = frame
        cv2.imshow(win_name, frame)

        key = cv2.waitKey(1 if streaming else 0) & 0xFF
        if key in (ord('q'), 27):
            break

    if not streaming and last_frame is not None:
        cv2.imshow(win_name, last_frame)
        cv2.waitKey(0)

    cv2.destroyWindow(win_name)

# ===== 사용 예시 =====
# ① 이미지 1장(URL 또는 로컬 경로)
#show_pose_window("https://ultralytics.com/images/bus.jpg")

# ③ 동영상 파일
show_pose_window("./images/pose_dance.mp4")
```

```python
from ultralytics import YOLO
import cv2, numpy as np, os

model = YOLO("./pose_dog.pt")

VIDEO_EXTS = {".mp4", ".avi", ".mov", ".mkv", ".webm"}

def is_video_source(src):
    if isinstance(src, int) or (isinstance(src, str) and src.isdigit()):
        return True
    if isinstance(src, str):
        ext = os.path.splitext(src)[1].lower()
        if ext in VIDEO_EXTS:
            return True
        if src.startswith(("rtsp://", "rtmp://")):
            return True
    return False

def show_with_indices(source, conf=0.25, imgsz=960, kp_thr=0.0):
    win = "Dog Pose (index helper)"
    cv2.namedWindow(win, cv2.WINDOW_NORMAL)
    streaming = is_video_source(source)
    paused = False
    last = None

    for r in model.predict(source=source, stream=True, conf=conf, imgsz=imgsz, vid_stride=1):
        if streaming and paused:
            frame = last.copy() if last is not None else r.orig_img.copy()
        else:
            frame = r.orig_img.copy()
            if r.keypoints is not None and r.keypoints.data is not None:
                k = r.keypoints.data
                try: k = k.cpu().numpy()
                except: pass
                if k.shape[0] > 0:
                    pts = k[0]
                    for idx, (x, y, c) in enumerate(pts):
                        if c > kp_thr:
                            cv2.circle(frame, (int(x), int(y)), 4, (0,255,255), -1)
                            cv2.putText(frame, str(idx), (int(x)+5, int(y)-5),
                                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2, cv2.LINE_AA)

        last = frame
        cv2.imshow(win, frame)

        wait = 1 if streaming else 1
        key = cv2.waitKey(wait) & 0xFF

        if key in (ord('q'), 27):
            break
        if streaming and key == ord(' '):
            paused = not paused

    if not streaming and last is not None:
        cv2.imshow(win, last)
        cv2.waitKey(0)

    cv2.destroyAllWindows()

# 사용 예시
# 이미지 1장
# show_with_indices("./images/pose_dog.jpg")

# 동영상 파일
show_with_indices("./images/pose_dog.mp4")
```

# Chapter 2. 객체 추적(Object Tracking)

![다중객체추적은 무엇인가?](./images/track1.gif)

![객체추적은 무엇인가?](./images/tracking2.gif)

#### 축구, 농구 등 스포츠 영상 속에서 각 선수들의 번호(id) 부여하고 선수들끼리 겹치거나 가려져도 번호(id)가 안 바뀌고 추적하는 것이 다중객체추적입니다.
--------------------------------------------------

#### MOT(다중 객체 추적) = 매 프레임에서 보이는 여러 물체를 찾아(검출) + 같은 물체끼리 연결해(추적) 같은 ID를 유지하는 것.

--------------------------------------
#### 지난주에 배운 객체 탐지만으로는 안되는가? (현실의 문제들)

#### 가림(occlusion): 사람이 잠깐 가려짐 → ID 잃기 쉬움

#### 비슷한 모습: 같은 유니폼/색 옷 → 서로 바뀌어 보임

#### 빠른/불규칙한 움직임: 축구 드리블, 갑자기 방향 전환

#### 카메라가 흔들림/움직임: CCTV 팬/틸트, 드론 영상

-------------------------------------------------------------
#### 주로 활용되는 분야는 다음과 같습니다

#### 자율주행: 도로의 차/사람을 계속 구분해서 추적

#### 스포츠 분석: 선수 위치/속도/패턴 분석

#### 보안/CCTV: 특정 사람 동선 파악

#### 로봇/드론: 주변 사물 지속 인식

```python
import cv2
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

video_path = './images/pose_dance.mp4'
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(frame, persist=True, classes=[0])

    annotated_frame = results[0].plot()

    cv2.imshow('YOLOv8 Tracking', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort

model = YOLO('yolov8n.pt')


tracker = DeepSort(max_age=60)

video_path = './images/pose_dance.mp4'
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, classes=[0])

    detections = []
    for result in results[0].boxes:
        x1, y1, x2, y2 = result.xyxy[0].tolist()
        confidence = result.conf[0].item()
        detections.append(([int(x1), int(y1), int(x2-x1), int(y2-y1)], confidence, 'person'))

    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:
        if not track.is_confirmed():
            continue
        track_id = track.track_id
        ltrb = track.to_ltrb()
        x1, y1, x2, y2 = map(int, ltrb)
        
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"Person ID: {track_id}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)


    cv2.imshow('DeepSORT Tracking', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Chapter 3. 이미지 분할(Image Segmentation)

![세그멘테이션이란 무엇인가?](./images/segmentation1.png)

![세그멘테이션 왜 쓰냐>?](./images/segmentation2.gif)

#### 이미지나 영상 속 픽셀 단위로 객체를 탐지하는 것이 image segmentation입니다.
#### 전통적으로 image segmentation은 세 종류로 나뉩니다.

![세그멘테이션의 3종류!](./images/segmentation3.png)

#### 1. Semantic segmentation
### 이미지 내 모든 팍실에 대해 객체 별로 분류 및 탐지를 한다.
### 배경(하늘, 땅 등)과 전경(사람, 자동차 등) 모두 진행

#### 2. Instance segmentation
### 전경(사람, 자동차 등)만 진행하며,
### 같은 객체라도 개별로 취급해서 분류 및 탐지한다.

#### 3. Panoptic segmentation
### 전경, 배경 모두 진행하며,
### 같은 객체라도 개별로 취급해서 분류 및 탐지한다.

#### 활용되는 사례는 다음과 같다:
#### 자율주행차: 도로/차선/보행자 위치를 정확히 구분
#### 의학영상: CT나 MRI에서 종양 부분 구분
#### 공장 검사: 제품 표면에서 결함 영역 구분
#### AR/필터: 사람만 분리해서 배경을 바꾸거나 이펙트 효과 부여

#### 이미지 파일 정하기

```python
import numpy as np, cv2, matplotlib.pyplot as plt
import torch, torchvision

def read_image_bgr(src):
    """src가 경로(str)든 배열이든 BGR 3채널 이미지로 반환."""
    if isinstance(src, str):
        if not os.path.exists(src):
            raise FileNotFoundError(f"경로가 존재하지 않습니다: {src}")
        data = np.fromfile(src, dtype=np.uint8)
        img = cv2.imdecode(data, cv2.IMREAD_COLOR)
        if img is None:
            img = cv2.imread(src, cv2.IMREAD_COLOR)
    elif isinstance(src, np.ndarray):
        img = src.copy()
    else:
        raise TypeError(f"user_img 타입이 이미지가 아닙니다: {type(src)}")

    if img is None:
        raise ValueError("이미지 로드 실패: 파일이 깨졌거나 형식이 지원되지 않습니다.")

    if img.ndim == 2:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    elif img.ndim == 3 and img.shape[2] == 4:
        img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)

    try:
        from PIL import Image, ImageOps
        if isinstance(src, str):
            with Image.open(src) as im:
                im = ImageOps.exif_transpose(im).convert("RGB")
                img = cv2.cvtColor(np.array(im), cv2.COLOR_RGB2BGR)
    except Exception:
        pass

    return img

def imshow_nb(title, img_bgr):
    """주피터에서 보기 좋은 표시 함수 (BGR->RGB)."""
    plt.figure()
    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
    plt.title(title); plt.axis("off"); plt.show()

# ===== 사용 예시 =====
user_img = "./images/segmentation1.jpg"
user_img = read_image_bgr(user_img)

print("OK:", user_img.shape, user_img.dtype)
imshow_nb("segmentation1.jpg", user_img)
```

```python
import numpy as np, cv2, matplotlib.pyplot as plt
import torch, torchvision
from torchvision.transforms.functional import to_tensor

def show(title, img, cmap=None):
    plt.figure()
    if img.ndim == 2:
        plt.imshow(img, cmap=cmap)
    else:
        # BGR->RGB for matplotlib
        if img.shape[2] == 3:
            import cv2 as _cv2
            plt.imshow(_cv2.cvtColor(img, _cv2.COLOR_BGR2RGB))
        else:
            plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

def fixed_color_for_id(i):
    rng = np.random.default_rng(12345)
    rng.bit_generator.state  # to silence linters
    # 간단한 해시 기반 고정 색
    r = (37 * i + 23) % 255
    g = (17 * i + 199) % 255
    b = (91 * i + 57) % 255
    return (int(b), int(g), int(r))  # BGR

def overlay_semantic_by_class(img_bgr, seg_pred, class_names=None, alpha=0.55):
    vis = img_bgr.copy()
    H, W = seg_pred.shape
    uniq = [c for c in np.unique(seg_pred) if c != 0]  # 0=background
    for cid in uniq:
        mask = (seg_pred == cid)
        color = fixed_color_for_id(int(cid))
        vis[mask] = (alpha*np.array(color) + (1-alpha)*vis[mask]).astype(np.uint8)
    # 범례(텍스트)
    if class_names is not None and len(uniq) > 0:
        plt.figure(); plt.imshow(cv2.cvtColor(vis, cv2.COLOR_BGR2RGB)); plt.axis('off')
        title = "Semantation — " + ", ".join([class_names[int(c)] for c in uniq if class_names and int(c) < len(class_names)])
        plt.title(title); plt.show()
    else:
        show("Semantic overlay", vis)
    return vis

def overlay_instances(img_bgr, masks_np, labels_np=None, scores_np=None, class_names=None, alpha=0.55, score_thr=0.5, min_area=500):
    """masks_np: [N,H,W] 0/1"""
    vis = img_bgr.copy()
    H, W = img_bgr.shape[:2]
    keep = []
    N = masks_np.shape[0]
    areas = masks_np.reshape(N, -1).sum(axis=1)
    for i in range(N):
        if scores_np is not None and scores_np[i] < score_thr: 
            continue
        if areas[i] < min_area: 
            continue
        keep.append(i)
    inst_id = 1
    for i in keep:
        mask = masks_np[i].astype(bool)
        color = fixed_color_for_id(inst_id + 1000)  # 인스턴스용 다른 시드
        vis[mask] = (alpha*np.array(color) + (1-alpha)*vis[mask]).astype(np.uint8)
        # 박스/라벨
        ys, xs = np.where(mask)
        if ys.size > 0:
            y1, x1, y2, x2 = ys.min(), xs.min(), ys.max(), xs.max()
            cv2.rectangle(vis, (x1,y1), (x2,y2), color, 2)
            txt = f"#{inst_id}"
            if labels_np is not None and class_names is not None:
                cid = int(labels_np[i])
                if 0 <= cid < len(class_names):
                    txt += f" {class_names[cid]}"
            if scores_np is not None:
                txt += f" {scores_np[i]:.2f}"
            cv2.putText(vis, txt, (x1, max(0,y1-5)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2, cv2.LINE_AA)
        inst_id += 1
    show(f"Instance overlay (kept={len(keep)})", vis)
    return vis


img_bgr = user_img.copy()
rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
H, W = img_bgr.shape[:2]
x = to_tensor(rgb).unsqueeze(0)
```

```python
from torchvision.models.segmentation import deeplabv3_resnet50, DeepLabV3_ResNet50_Weights
seg_weights = DeepLabV3_ResNet50_Weights.DEFAULT
seg_model = deeplabv3_resnet50(weights=seg_weights).eval()
with torch.no_grad():
    out = seg_model(x)
seg_pred = out['out'][0].argmax(0).cpu().numpy().astype(np.uint8)

sem_class_names = seg_weights.meta.get("categories", None)
_ = overlay_semantic_by_class(img_bgr, seg_pred, class_names=sem_class_names, alpha=0.6)
```

```python
from torchvision.models.detection import maskrcnn_resnet50_fpn, MaskRCNN_ResNet50_FPN_Weights
det_weights = MaskRCNN_ResNet50_FPN_Weights.DEFAULT
det_model = maskrcnn_resnet50_fpn(weights=det_weights).eval()
with torch.no_grad():
    det_out = det_model([x[0]])[0]

scores = det_out['scores'].cpu().numpy()
labels = det_out['labels'].cpu().numpy()
masks  = (det_out['masks'][:,0].cpu().numpy() > 0.5).astype(np.uint8)

det_class_names = det_weights.meta.get("categories", None)
_ = overlay_instances(img_bgr, masks, labels_np=labels, scores_np=scores, class_names=det_class_names, alpha=0.6, score_thr=0.6, min_area=500)
```
